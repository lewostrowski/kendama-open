```python
def groupGeoProg(tableName):
    geometricalProgression = []
    start = 2
    ratio = 2
    while start <= len(tableName):
        geometricalProgression.append(start)
        start = start * ratio
    return geometricalProgression
```

```python
def groupCreator(tableName, gPopulation=0, gNumber=0, strict=False):   
    if gPopulation == 0 and gNumber > 0: forceGroup=True
    elif gPopulation > 0 and gNumber == 0: forceGroup=False
    
    if gPopulation < 0 and gPopulation > len(playerArray) / gPopulation: return('Wrong group population')
    
    
    playerArray = groupPlayerArray(tableName)
    orginalArray = playerArray.copy()
    groups = []
    if gPopulation > 0 and gNumber > 0 and len(tableName) % gNumber == 0: 
        return('Forced group number caused error with forced gorup population.')
    elif gNumber > 0 and gPopulation == 0: gPopulation = math.ceil(len(tableName)/gNumber)
    
    if len(orginalArray) % gPopulation == 0:
        while len(groups) < len(orginalArray)/gPopulation:
            group = []
            while len(group) < gPopulation:
                group.append(playerArray[0])
                playerArray.pop(0)
            groups.append(group)
        return groups 
    else:
        playerIndex = 0
        while len(groups) < math.floor(len(orginalArray) / gPopulation):
            group = []
            while len(group) < gPopulation:
                group.append(playerArray[playerIndex])
                playerArray.pop(playerIndex)
                
            groups.append(group)    
        
        groups.append(playerArray)
        
        if len(groups[-1]) < len(groups):
            if forceGroup == True: 
                for p in groups[-1]: groups[groups[-1].index(p)].append(p)
                groups.pop(-1)
            elif forceGroup == False and strict == False: 
                for p in groups[-1]: groups[groups[-1].index(p)].append(p)
                groups.pop(-1)
            else: return print('Forced strict mode caused error.')
        elif len(groups[-1]) > len(groups):
            if forceGroup == True and strict == False:
                pLeft = 0
                while len(groups[-1]) != pLeft:
                    pIndex = 0
                    for p in groups[-1]:
                        if pIndex < len(groups[-1]):
                            groups[pIndex].append(p)
                            pIndex += 1
                            pLeft += 1
                        else: 
                            pIndex = 0
                            groups[pIndex].append(p)
                            pIndex += 1
                            pLeft += 1
            if forceGroup == False and strict == False:
                if len(groups[-1]) >= len(groups[0])*0.75:
                    pass
                else: return print('Forced strict mode caused error.')
                
        return groups
```

```python
def chooseBestGroups(tableName, minP=0,maxP=0):
    playerArray = groupPlayerArray(tableName)
    
    if minP == 0 and maxP == 0:
        minP = 3
        maxP = 6
        if len(playerArray) > 50:
            maxP = 10
    elif minP > 0 and maxP == 0:
        maxP = minP*2
    elif minP == 0 and maxP > 0:
        minP = math.ceil(maxP/2)
        
    count = minP
    avalibleGroupPop = []
    while count <= maxP:
        avalibleGroupPop.append(count)
        count += 1

    avalibleEqualSchema = []
    for n in avalibleGroupPop:
        if len(playerArray) % n == 0:
            avalibleEqualSchema.append(n)

    if len(avalibleEqualSchema) == 1:
        return groupCreator(tableName, gPopulation=avalibleEqualSchema[0])
    elif len(avalibleEqualSchema) > 1:
        if len(playerArray)/maxP > maxP:
            return groupCreator(tableName, gPopulation=avalibleEqualSchema[-1])
        else: return groupCreator(tableName, gPopulation=avalibleEqualSchema[1])
    elif len(avalibleEqualSchema) == 0:
        if len(avalibleGroupPop) == 1: return groupCreator(tableName, gPopulation=avalibleGroupPop[0])
        elif len(avalibleGroupPop) == 2:
            if len(playerArray)/maxP > maxP:
                return groupCreator(tableName, gPopulation=avalibleGroupPop[-1])
            else: return groupCreator(tableName, gPopulation=avalibleGroupPop[1])
        else: return groupCreator(tableName, gPopulation=avalibleGroupPop[-2])

    return print('end of function')
```

```python
def groupCreate(tableName, forceLadder=False, forceGroupNumber=0, forceGroupPopulation=0, forceStrict=False, setMinGroupPopulation=0, setMaxGroupPopulation=0):
    
    #CHECK TABLE AND CONDITION
    if forceGroupNumber > 0 or forceGroupNumber > 0:
        if (forceGroupPopulation < 2 or forceGroupPopulation > len(tableName) / 2) and forceGroupPopulation != 0:
            return print('Forced group population caused error.')
        elif (forceGroupNumber > len(tableName) / 2 or forceGroupNumber < 2) and forceGroupNumber != 0:
            return print('Forced group number caused error.')
        elif len(tableName) / forceGroupPopulation != forceGroupNumber and (forceGroupPopulation > 0 and forceGroupNumber > 0):
            return print('Forced group number caused error with forced gorup population.')
    
    geometricalProgression = groupGeoProg(tableName)
    
    #CREAT LADDER WITH PLAYERS
    if forceLadder == True and len(tableName) in geometricalProgression and len(tableName) >= 4: 
        gameArray = groupMakeLadder(tableName, geometricalProgression)
        
    #CREATE GROUPS WITH FORCED GROUP NUMBER
    elif forceGroupNumber > 0:  
        if forceStrict == True:    
            if len(tableName) % forceGroupNumber == 0: 
                gameArray = groupCreator(tableName, gPopulation=len(tableName)/forceGroupNumber, strict=True)       
            else: return print('forceStrict function prevents group maker.')
        else:
            gP = math.ceil(len(tableName)/forceGroupNumber)
            gameArray = groupCreator(tableName, gPopulation=gP)
    
    #CREATE GROUPS WITH FORCED GROUP POPULATION
    elif forceGroupPopulation > 0:
        if forceStrict == True: 
            if len(tableName) % forceGroupPopulation == 0:
                gameArray = groupCreator(tableName, gPopulation=forceGroupPopulation, strict=True) 
            else: return print('forceStrict function prevents group maker.')
        else: gameArray = groupCreator(tableName, gPopulation=forceGroupPopulation) 
            
    #CREATE RANDOM GROUPS
    elif forceGroupNumber == 0 and forceGroupPopulation == 0:
        gameArray = chooseBestGroups(tableName, minP=setMinGroupPopulation, maxP=setMaxGroupPopulation)
    
    return gameArray
```
